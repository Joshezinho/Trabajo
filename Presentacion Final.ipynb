Hay investigadores, como Ouviña (2012) que señalan que los actores involucrados en manifestaciones como la del 2006 o la del 2011 pertenecen a una generación que perdió el miedo a protestar puesto que no nacieron en dictadura. Al mismo tiempo los autores [@sandovalmoyaGeneracionSinMiedo2019] proponen que los jóvenes han resignificado la política en un desarrollo particular que cae bajo el marco de la postdictadura y al mismo tiempo, estos mismos jovenes han experimentado un crecimiento sustancial en materia política a partir de las mismas movilizaciones estudiantiles.


# Introducción
Los movimientos sociales en general se ven como una acumulación de malestar, pero @monterobarrigaESTRATEGIASRECURSOSEMPLEADOS2017, exponen desde esta perspectiva que los movimientos sociales serían también una opción política, razonada y evaluada por sus integrantes como un acto necesario y útil en pro de conseguir ciertos objetivos. Pero, estos mismos movimientos no sólo generan gran comprensión y adhesión por parte de la ciudadanía por ser disruptivos y agrupar amplios sectores, sino también por la era del internet según lo expuesto por @rodriguezFormasMediosComunicacion2018.

### Hipóteis

- El apoyo al movimiento estudiantil no responde a un atributo como el sexo.
- Las personas más jóvenes son las que más apoyan al movimiento estudiantil.
- Mientras mayor el nivel educativo, mayor será el apoyo.
